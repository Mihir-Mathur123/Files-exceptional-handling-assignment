Q.15)How does Python's garbage collection system work?--

âœ… How Python's Garbage Collection System Works
Python uses an automatic garbage collection system to manage memory and ensure efficient usage of resources during the program's execution. The garbage collector is responsible for identifying and freeing memory occupied by objects that are no longer needed, thus preventing memory leaks and improving performance.

Letâ€™s break down how Pythonâ€™s garbage collection works.

ðŸ”¹ 1. Reference Counting (Primary Method)
Python uses a reference counting technique as the primary mechanism for managing memory. Each object in Python has an associated reference count, which tracks how many references point to that object.

Reference Count: Every time an object is referenced (i.e., assigned to a variable or passed as a function argument), its reference count is incremented. When the reference goes out of scope or is deleted, the count is decremented.

Example:


```python
import sys

a = [1, 2, 3]  # Reference count for the list object increases by 1
b = a  # Reference count for the list object increases by 1
print(sys.getrefcount(a))  # Output: 3 (a, b, and sys.getrefcount refer to the same object)
del a  # Reference count for the list object decreases by 1
print(sys.getrefcount(b))  # Output: 2 (b and sys.getrefcount still refer to the object)
```

When the reference count reaches zero, meaning no references to the object exist anymore, Python automatically deallocates the memory occupied by the object.

ðŸ”¹ 2. Problems with Reference Counting
While reference counting is simple and efficient, it has some limitations, the most notable being the problem of circular references.

Example of Circular References:

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

# Creating circular reference
node1 = Node(1)
node2 = Node(2)
node1.next = node2
node2.next = node1
```

In this case, both node1 and node2 reference each other. When the program goes out of scope, reference counting will not be able to detect that these objects are no longer needed because each still holds a reference to the other.

This leads to memory leaks since Python won't be able to automatically deallocate their memory.

Q.25)Why is it important to use multithreading in certain applications?--

âœ… Why is it Important to Use Multithreading in Certain Applications?
Multithreading is a powerful concept in concurrent programming where a single process is divided into multiple threads of execution. Each thread runs concurrently, allowing different parts of the program to execute simultaneously or in overlapping intervals. This can significantly improve the performance of certain types of applications, especially when they involve tasks that can be executed in parallel or require non-blocking operations.

Letâ€™s explore the importance of multithreading and the specific scenarios where it's particularly beneficial:

ðŸ”¹ 1. Improved Performance for I/O-bound Tasks
One of the most common reasons to use multithreading is to improve performance in I/O-bound tasks. These tasks typically involve waiting for external resources, such as reading from disk, network requests, or waiting for user input. Multithreading helps by allowing other threads to run while one thread is waiting for I/O operations to complete.

I/O-bound tasks: These tasks involve waiting for input/output operations to finish, such as reading from files, making network requests, or interacting with databases.

Example:

```python
import threading
import time

def download_data():
    print("Start downloading...")
    time.sleep(3)  # Simulating a network request
    print("Download complete!")

def process_data():
    print("Start processing...")
    time.sleep(2)  # Simulating a CPU-bound task
    print("Processing complete!")

# Without multithreading, we would have to wait for one to finish before starting the other
# With multithreading, both tasks can run concurrently
thread1 = threading.Thread(target=download_data)
thread2 = threading.Thread(target=process_data)

thread1.start()
thread2.start()

thread1.join()
thread2.join()
```

In this example:

download_data() simulates waiting for a network request.

process_data() simulates a processing task that could run concurrently with the download.

Without multithreading, you'd have to wait for the download to finish before starting the processing. With multithreading, both tasks can overlap, improving overall efficiency.

2. Responsiveness in User Interface (UI) Applications
In applications with a graphical user interface (GUI), itâ€™s essential to keep the interface responsive while executing long-running tasks. Multithreading allows you to run background tasks without blocking the main thread, which is responsible for handling user input and updating the UI.

For example, in a GUI application (like a desktop app built with Tkinter or PyQt), long-running tasks like downloading data, file processing, or image manipulation can be offloaded to separate threads to prevent the UI from freezing.

Example: Using tkinter and Multithreading for Responsiveness

ðŸ”¹ 3. Handling Concurrent Requests in Web Servers
In web servers or network applications, multithreading is crucial to handle multiple client requests concurrently. Without multithreading, the server would need to process requests sequentially, which can be inefficient and lead to long wait times for clients.

Using multithreading, each incoming request can be handled by a separate thread, improving the responsiveness of the server and allowing it to serve many clients simultaneously.

```python
def get_dictionary_value(my_dict, key):
    """
    Attempts to retrieve a value from a dictionary using a specified key.
    Uses a try-except block to handle KeyError if the key does not exist.

    Args:
        my_dict (dict): The dictionary to search in.
        key: The key to look for.

    Returns:
        The value associated with the key if found, otherwise None.
    """
    try:
        value = my_dict[key]
        print(f"Key '{key}' found. Value: {value}")
        return value
    except KeyError:
        print(f"Error: Key '{key}' not found in the dictionary.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# --- Test Cases ---

# 1. Dictionary with various data
my_data = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "occupation": "Engineer"
}

print("--- Test Case 1: Existing Keys ---")
get_dictionary_value(my_data, "name")
get_dictionary_value(my_data, "age")
get_dictionary_value(my_data, "city")

print("\n--- Test Case 2: Non-Existing Key ---")
get_dictionary_value(my_data, "country") # This key does not exist
get_dictionary_value(my_data, "zip_code") # This key also does not exist

print("\n--- Test Case 3: Empty Dictionary ---")
empty_dict = {}
get_dictionary_value(empty_dict, "item")

print("\n--- Test Case 4: Key with None value (different from missing key) ---")
dict_with_none = {"status": None, "id": 123}
get_dictionary_value(dict_with_none, "status") # Key exists, value is None
get_dictionary_value(dict_with_none, "description") # Key does not exist
```

Q.13)How would you check if a file exists before attempting to read it in Python?

In Python, there are a few robust ways to check if a file exists before attempting to read it. This is good practice to prevent FileNotFoundError and make your program more resilient.

Here are the most common and recommended methods:

Method 1: Using os.path.exists() or os.path.isfile()
The os module provides functions for interacting with the operating system, including file system operations.

os.path.exists(path): Checks if a path (which can be a file or a directory) exists. Returns True if it exists, False otherwise.
os.path.isfile(path): Checks if a path exists and is a regular file. Returns True if it's an existing file, False otherwise (e.g., if it's a directory or doesn't exist). This is generally preferred if you specifically want to read a file.

```python
import os

file_path = "my_document.txt"

# Create a dummy file for demonstration (optional)
with open(file_path, 'w') as f:
    f.write("Hello, this is a test file.")

print(f"Checking for file: '{file_path}'")

# Using os.path.exists()
if os.path.exists(file_path):
    print("Method 1 (os.path.exists): File exists!")
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print("File content:\n", content)
    except Exception as e:
        print(f"An error occurred while reading: {e}")
else:
    print("Method 1 (os.path.exists): File does not exist.")

print("-" * 30)

# Using os.path.isfile() - more specific for files
if os.path.isfile(file_path):
    print("Method 1 (os.path.isfile): File exists and is a regular file!")
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print("File content:\n", content)
    except Exception as e:
        print(f"An error occurred while reading: {e}")
else:
    print("Method 1 (os.path.isfile): File does not exist or is not a regular file.")

# Test with a non-existent file
non_existent_file = "non_existent.txt"
print(f"\nChecking for non-existent file: '{non_existent_file}'")
if os.path.isfile(non_existent_file):
    print(f"File '{non_existent_file}' exists.")
else:
    print(f"File '{non_existent_file}' does not exist.")

# Clean up the dummy file (optional)
if os.path.exists(file_path):
    os.remove(file_path)
```

Method 2: Using pathlib (Modern Pythonic Way - Python 3.4+)
The pathlib module provides an object-oriented way to interact with file system paths, which is often considered more intuitive and powerful than os.path.

```python
from pathlib import Path

file_path_str = "another_document.txt"
file_path = Path(file_path_str) # Create a Path object

# Create a dummy file for demonstration (optional)
with open(file_path, 'w') as f:
    f.write("This is another test document.")

print(f"\nChecking for file: '{file_path_str}' using pathlib")

if file_path.exists():
    print("Method 2 (pathlib.Path.exists): File exists!")
    if file_path.is_file(): # Check if it's specifically a file
        print("Method 2 (pathlib.Path.is_file): It is indeed a file!")
        try:
            with open(file_path, 'r') as file:
                content = file.read()
                print("File content:\n", content)
        except Exception as e:
            print(f"An error occurred while reading: {e}")
    else:
        print("Method 2 (pathlib.Path.is_file): It exists, but is not a regular file (e.g., it's a directory).")
else:
    print("Method 2 (pathlib.Path.exists): File does not exist.")

# Test with a non-existent file
non_existent_file_path = Path("yet_another_non_existent.txt")
print(f"\nChecking for non-existent file: '{non_existent_file_path}'")
if non_existent_file_path.is_file():
    print(f"File '{non_existent_file_path}' exists.")
else:
    print(f"File '{non_existent_file_path}' does not exist.")

# Clean up the dummy file (optional)
if file_path.exists():
    file_path.unlink() # For pathlib, use unlink() to delete a file
```

Q.16)Demonstrate how to use memory profiling to check the memory usage of a small program.


Method 1: Using tracemalloc (Built-in)
tracemalloc allows you to track memory allocations and see a traceback of where specific memory blocks were allocated.

Installation: No installation needed,

```python
import tracemalloc
import sys
import gc # Garbage Collector

def demonstrate_tracemalloc():
    print("--- Demonstrating tracemalloc for Memory Profiling ---")

    # 1. Start tracking memory allocations
    tracemalloc.start()
    print("tracemalloc started.")

    # Get initial memory usage
    initial_current, initial_peak = tracemalloc.get_traced_memory()
    print(f"Initial memory usage: Current={initial_current / 1024:.2f} KB, Peak={initial_peak / 1024:.2f} KB")

    # --- Small program logic ---
    my_list = []
    num_elements = 100000

    print(f"\nAllocating {num_elements} integers into a list...")
    for i in range(num_elements):
        my_list.append(i) # Appending integers

    # Get memory usage after allocation
    after_alloc_current, after_alloc_peak = tracemalloc.get_traced_memory()
    print(f"After allocation: Current={after_alloc_current / 1024:.2f} KB, Peak={after_alloc_peak / 1024:.2f} KB")
    print(f"Memory increased by: {(after_alloc_current - initial_current) / 1024:.2f} KB")

    # Let's create another data structure
    my_dict = {f"key_{i}": f"value_{i}" for i in range(50000)}
    print(f"\nAllocating {len(my_dict)} key-value pairs into a dictionary...")

    # Get memory usage after another allocation
    final_current, final_peak = tracemalloc.get_traced_memory()
    print(f"After dictionary allocation: Current={final_current / 1024:.2f} KB, Peak={final_peak / 1024:.2f} KB")
    print(f"Memory increased by (since start): {(final_current - initial_current) / 1024:.2f} KB")


    # --- Analyze snapshots ---
    print("\n--- Top 10 Memory Allocations (by file and line) ---")
    # Take a snapshot of current allocations
    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno') # Group by line number

    for stat in top_stats[:10]:
        print(stat)

    # Optional: Analyze differences between snapshots
    print("\n--- Memory Differences (Initial vs. Final) ---")
    initial_snapshot = tracemalloc.take_snapshot()
    # Perform some operations that might release memory (e.g., delete large objects)
    del my_list
    del my_dict
    gc.collect() # Force garbage collection

    final_snapshot_after_del = tracemalloc.take_snapshot()
    top_diffs = final_snapshot_after_del.compare_to(initial_snapshot, 'lineno')

    # Note: Differences might be tricky as Python's GC can reclaim memory at various times.
    # For more reliable diffs, run the code segment you want to profile between `start()` and `stop()`.

    print("\n--- Top 10 Memory Differences (after object deletion) ---")
    for stat in top_diffs[:10]:
        print(stat)


    # 2. Stop tracking memory allocations
    tracemalloc.stop()
    print("\ntracemalloc stopped.")
    print(f"Final memory state (after stop): Current={tracemalloc.get_traced_memory()[0] / 1024:.2f} KB")


if __name__ == "__main__":
    demonstrate_tracemalloc()
```

Q.21)Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
import os
import re # Import the regular expression module for word boundary handling

def count_word_occurrences(file_path, target_word, case_sensitive=False):
    """
    Reads a text file and counts the occurrences of a specific word.

    Args:
        file_path (str): The path to the text file.
        target_word (str): The word to search for.
        case_sensitive (bool): If True, the search is case-sensitive.
                               If False, the search is case-insensitive.
                               Defaults to False.

    Returns:
        int: The number of occurrences of the target word.
             Returns -1 if the file is not found or an error occurs.
    """
    if not os.path.exists(file_path):
        print(f"Error: The file '{file_path}' does not exist.")
        return -1

    count = 0
    search_word = target_word
    if not case_sensitive:
        search_word = target_word.lower() # Convert target word to lowercase for case-insensitive comparison

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                # For robust word counting, we often want to match whole words.
                # We can convert the line to lowercase here for case-insensitive matching.
                processed_line = line
                if not case_sensitive:
                    processed_line = line.lower()

                # Option 1: Simple string count (may count substrings)
                # count += processed_line.count(search_word)

                # Option 2: Using regex for whole word matching (recommended)
                # \b matches a word boundary (e.g., prevents "apple" from matching "pineapple")
                # re.findall returns a list of all non-overlapping matches
                pattern = r'\b' + re.escape(search_word) + r'\b' # re.escape handles special regex chars in target_word
                matches = re.findall(pattern, processed_line)
                count += len(matches)

        return count

    except FileNotFoundError: # This is technically caught by os.path.exists, but good to have
        print(f"Error: The file '{file_path}' was not found.")
        return -1
    except Exception as e:
        print(f"An unexpected error occurred while reading the file: {e}")
        return -1

# --- Main Program Execution ---
if __name__ == "__main__":
    # 1. Create a dummy text file for demonstration
    file_name = "sample_text.txt"
    content = """
    Python is a popular programming language.
    It is widely used for web development, data analysis, machine learning, and more.
    Many Python developers find Python code to be clean and readable.
    The Python community is very supportive.
    Let's count how many times "python" appears, and "count".
    This is also about COUNTING.
    """
    with open(file_name, 'w', encoding='utf-8') as f:
        f.write(content.strip()) # .strip() removes leading/trailing whitespace

    print(f"Created '{file_name}' for demonstration.")

    # 2. Define the word to search for
    word_to_find = "Python"

    # 3. Test Case 1: Case-insensitive search
    print(f"\n--- Searching for '{word_to_find}' (case-insensitive) ---")
    occurrences_ci = count_word_occurrences(file_name, word_to_find, case_sensitive=False)
    if occurrences_ci != -1:
        print(f"The word '{word_to_find}' appears {occurrences_ci} times (case-insensitive).")

    # Expected: 5 (Python, Python, Python, python, Python)

    # 4. Test Case 2: Case-sensitive search
    print(f"\n--- Searching for '{word_to_find}' (case-sensitive) ---")
    occurrences_cs = count_word_occurrences(file_name, word_to_find, case_sensitive=True)
    if occurrences_cs != -1:
        print(f"The word '{word_to_find}' appears {occurrences_cs} times (case-sensitive).")
    # Expected: 4 (Python, Python, Python, Python)

    # 5. Test Case 3: Another word, case-insensitive (and demonstrating word boundaries)
    word_to_find_2 = "count"
    print(f"\n--- Searching for '{word_to_find_2}' (case-insensitive, whole word) ---")
    occurrences_count = count_word_occurrences(file_name, word_to_find_2, case_sensitive=False)
    if occurrences_count != -1:
        print(f"The word '{word_to_find_2}' appears {occurrences_count} times (case-insensitive, whole word).")
    # Expected: 2 (count, COUNTING) - should not match "account" if present in file

    # 6. Test Case 4: Non-existent file
    print(f"\n--- Searching in a non-existent file ---")
    count_word_occurrences("non_existent_file.txt", "test")

    # --- Clean up the dummy file ---
    if os.path.exists(file_name):
        os.remove(file_name)
        print(f"\nCleaned up '{file_name}'.")
```

Q.22)How can you check if a file is empty before attempting to read its contents?

```python
import os
from pathlib import Path

def is_file_empty_getsize(file_path):
    """
    Checks if a file is empty using os.path.getsize().
    Returns True if empty, False if not empty, None if file doesn't exist or error.
    """
    try:
        if not os.path.exists(file_path):
            print(f"'{file_path}': File does not exist.")
            return None

        file_size = os.path.getsize(file_path)
        return file_size == 0
    except OSError as e:
        print(f"Error checking size of '{file_path}': {e}")
        return None

def is_file_empty_read_char(file_path):
    """
    Checks if a file is empty by attempting to read its first character.
    Returns True if empty, False if not empty, None if file doesn't exist or error.
    """
    try:
        if not os.path.exists(file_path):
            print(f"'{file_path}': File does not exist.")
            return None

        with open(file_path, 'r', encoding='utf-8') as file:
            first_char = file.read(1) # Read only 1 character
            return not bool(first_char) # If an empty string is returned, it's empty
    except FileNotFoundError: # Caught by os.path.exists, but good for robustness
        print(f"'{file_path}': File not found.")
        return None
    except Exception as e:
        print(f"Error reading first char of '{file_path}': {e}")
        return None

def is_file_empty_pathlib(file_path):
    """
    Checks if a file is empty using pathlib.Path.stat().st_size.
    Returns True if empty, False if not empty, None if file doesn't exist or error.
    """
    try:
        path_obj = Path(file_path)
        if not path_obj.exists():
            print(f"'{file_path}': File does not exist.")
            return None

        return path_obj.stat().st_size == 0
    except OSError as e:
        print(f"Error checking size of '{file_path}' with pathlib: {e}")
        return None


def process_file(file_path):
    """
    Demonstrates how to process a file only if it's not empty.
    Uses os.path.getsize() for the check.
    """
    print(f"\n--- Processing '{file_path}' ---")

    is_empty = is_file_empty_getsize(file_path) # Choose your preferred method here

    if is_empty is None: # An error occurred or file doesn't exist
        print(f"Cannot process '{file_path}' due to an error or non-existence.")
        return
    elif is_empty:
        print(f"'{file_path}' is empty. Skipping content reading.")
    else:
        print(f"'{file_path}' is not empty. Reading content...")
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
                print("Content:")
                print(content)
        except Exception as e:
            print(f"An error occurred while reading content of '{file_path}': {e}")


# --- Main Demonstration ---
if __name__ == "__main__":
    # Create dummy files for testing
    empty_file_name = "empty_test.txt"
    non_empty_file_name = "non_empty_test.txt"
    non_existent_file_name = "does_not_exist.txt"

    # Create an empty file
    open(empty_file_name, 'w').close()
    print(f"Created empty file: '{empty_file_name}'")

    # Create a non-empty file
    with open(non_empty_file_name, 'w', encoding='utf-8') as f:
        f.write("Hello, World!\n")
        f.write("This file has some content.")
    print(f"Created non-empty file: '{non_empty_file_name}'")

    print("\n--- Using os.path.getsize() ---")
    print(f"Is '{empty_file_name}' empty? {is_file_empty_getsize(empty_file_name)}")
    print(f"Is '{non_empty_file_name}' empty? {is_file_empty_getsize(non_empty_file_name)}")
    print(f"Is '{non_existent_file_name}' empty? {is_file_empty_getsize(non_existent_file_name)}")

    print("\n--- Using read(1) ---")
    print(f"Is '{empty_file_name}' empty? {is_file_empty_read_char(empty_file_name)}")
    print(f"Is '{non_empty_file_name}' empty? {is_file_empty_read_char(non_empty_file_name)}")
    print(f"Is '{non_existent_file_name}' empty? {is_file_empty_read_char(non_existent_file_name)}")

    print("\n--- Using pathlib.Path.stat().st_size ---")
    print(f"Is '{empty_file_name}' empty? {is_file_empty_pathlib(empty_file_name)}")
    print(f"Is '{non_empty_file_name}' empty? {is_file_empty_pathlib(non_empty_file_name)}")
    print(f"Is '{non_existent_file_name}' empty? {is_file_empty_pathlib(non_existent_file_name)}")


    # Demonstrate actual file processing flow
    process_file(empty_file_name)
    process_file(non_empty_file_name)
    process_file(non_existent_file_name)

    # Clean up dummy files
    print("\n--- Cleaning up files ---")
    if os.path.exists(empty_file_name):
        os.remove(empty_file_name)
        print(f"Removed '{empty_file_name}'")
    if os.path.exists(non_empty_file_name):
        os.remove(non_empty_file_name)
        print(f"Removed '{non_empty_file_name}'")
```

Q.23)Write a Python program that writes to a log file when an error occurs during file handling.

```python
import logging

# Configure logging to write to a file with the level ERROR
logging.basicConfig(filename='error_log.txt',
                    level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def read_file(filename):
    try:
        with open(filename, 'r') as file:
            contents = file.read()
            print(contents)
    except (FileNotFoundError, IOError) as e:
        logging.error(f"Error occurred while handling the file '{filename}': {e}")
        print("An error occurred. Please check the log file for details.")

# Example usage
read_file('non_existent_file.txt')
```
